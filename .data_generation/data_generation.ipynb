```python
import random

Forme = ["Boule","Pyramide"]
P1 = ["lisse","Pointe"]
P2 = ["Chaude","Froide"]
P3 = ["Blanc","Bleu"]
P4 = ["Creuse","Pleine"]
P5 = ["10 cm","12 cm"]

Liste = open("drive/MyDrive/codeWeek/armes.doc", "w")
ListeArmeSans = []
ListeArmeAvec = []

for i in Forme :
    for j in P1 :
        for k in P2 :
            for e in P3 :
                for f in P4 :
                    for s in P5 :
                        ListeArmeSans.append(i+' ('+j+','+k+','+e+','+f+','+s+')')

for arme1 in ListeArmeSans: 
  for arme2 in ListeArmeSans:
    if not (arme1==arme2):
      ListeArmeAvec.append(arme1 + " " + arme2+'\n')

random.shuffle(ListeArmeAvec)

for index in range(0, len(ListeArmeAvec)):
  Liste.writelines(ListeArmeAvec[index])

print(len(ListeArmeAvec))
Liste.close()
```
